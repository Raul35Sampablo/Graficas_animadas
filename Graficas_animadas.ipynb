```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
t = np.linspace(0,10,100)                      #Eje X
f0 = 1 
s = np.cos( 2 * np.pi * f0 * t - np.pi/2 )   #Eje Y
s2 = np.cos( 2 * np.pi * 10 * t - np.pi/2 )   #Eje Y
s3 = np.random.uniform( low = -1, high = 1, size = len(t) )
```

```python
def animacion(x, y):

  fig, ax = plt.subplots()
  line, = ax.plot(x,y)
  num_frames = len(x)

  def init():
    #Seleccionamos el punto de inicio el cual comienza 0,0
    line.set_data([0], [0])
    return line,

  def update(frame):
    mask = np.zeros_like(t, dtype=bool)   #Creamos una "array mascara" con todos los valores iniciales en cero
    mask[:frame + 1] = True               #Convertimos todos a true en cada iteraccion
    x_data = x[mask]
    y_data = y[mask]
    line.set_data(x_data, y_data)
    return line,

  ani = animation.FuncAnimation(fig, update, frames = num_frames, init_func=init, blit=True, interval=50)
  plt.close(fig)
  
  return HTML(ani.to_jshtml())
```

```python
res = animacion(t, s2)
res
```
